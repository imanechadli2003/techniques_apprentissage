```python
# Chargement et affichage des premières données
import pandas as pd

train_data = pd.read_csv('../data/raw/extracted/train.csv')
test_data  = pd.read_csv('../data/raw/extracted/test.csv')

train_data.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))

species = train_data['species'].unique()

for sp in species:
    subset = train_data[train_data['species'] == sp]
    plt.scatter(
        subset['shape1'],
        subset['margin1'],
        label=sp,
        alpha=0.6,
        s=15
    )

plt.xlabel("shape1")
plt.ylabel("margin1")
plt.title("shape1 vs margin1")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
